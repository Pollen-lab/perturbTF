# Individual assignment

Assigning individual based on metadata of 10X pooling and SNP genotyping from scRNA-seq data derived from Vireo

```python
import numpy as np
import matplotlib.pyplot as plt
import vireoSNP
import os
```

## 2D

```python
directory='/wynton/scratch/jding/vireo'
```

```python
f1 = '2D_L2'
f2 = '2D_L3'
res = vireoSNP.vcf.match_VCF_samples(os.path.join(directory,f1,'GT_donors.vireo.vcf.gz'),
                                     os.path.join(directory,f2,'GT_donors.vireo.vcf.gz'),
                                     GT_tag1 = 'PL', GT_tag2='PL')

fig = plt.figure()
vireoSNP.plot.heat_matrix(res['matched_GPb_diff'], 
                          res['matched_donors1'] , 
                          res['matched_donors2'] )
plt.title("Geno Prob Delta: %d SNPs" %(res['matched_n_var']))
plt.tight_layout()
plt.show()
```
